```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load the iris dataset as an example
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Create a decision tree classifier
clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train, y_train)

# Prune the tree using the leaf bound
def prune_tree_leaf_bound(tree, min_leaves):
    n_nodes = tree.node_count
    children_left = tree.children_left
    children_right = tree.children_right

    # Create a list of nodes to remove
    nodes_to_remove = []
    for i in range(n_nodes):
        if children_left[i] == children_right[i]:
            n_leaves = 1
            for j in (children_left[i], children_right[i]):
                if j != -1:
                    n_leaves += 1
            if n_leaves <= min_leaves:
                nodes_to_remove.append(i)

    # Remove the nodes
    for i in nodes_to_remove:
        children_left[i] = -1
        children_right[i] = -1

prune_tree_leaf_bound(clf.tree_, 2)

# evaluate the pruned tree on the test set
accuracy = clf.score(X_test, y_test)
print("Accuracy of the pruned tree: {:.2f}%".format(accuracy*100))
```

```python
!pip install transformers
import re
import string
from itertools import chain

import pandas as pd
from transformers import BertTokenizer

# Load the pre-trained model tokenizer (vocabulary)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)

def clean_text(text: str) -> str:
    # Use a regular expression to remove punctuation, URLs, emails, and numbers
    text = re.sub(f'[{string.punctuation}]', '', text)
    text = re.sub(r'www\S+', '', text)
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'\S*@\S*\s?', '', text)
    text = ''.join([i for i in text if not i.isdigit()])
    return text.lower()

# Use a generator function to yield the data one batch at a time
def process_text(file):
    with open(file, encoding='utf-8-sig') as f:
        # Process each line of the file
        for line in f:
            # Clean the text and yield the result
            yield tokenizer.tokenize(clean_text(line))

# Use the map function to apply the process_text function to each line of the file
text_tokens = map(process_text, ['bible.txt'])

# Use the itertools module to flatten the list of lists
text_tokens = list(chain.from_iterable(text_tokens))

# Create a Pandas DataFrame from the list of tokens
df = pd.DataFrame({'tokens': text_tokens})
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load the iris dataset as an example
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Create a decision tree classifier
clf = DecisionTreeClassifier(random_state=0)

print("*************")
# Fit the model
print("Fitting the model")
clf.fit(X_train, y_train)
print(X_train, y_train)
# Prune the tree using the leaf bound
def prune_tree_leaf_bound(tree, min_leaves):
    n_nodes = tree.node_count
    children_left = tree.children_left
    children_right = tree.children_right
    # print n_nodes = tree.node_count
    print("*************")
    print("n_nodes = tree.node_count")
    print(n_nodes)
    print("*************")
    print("children_left = tree.children_left")
    print(children_left)
    print("*************")
    print("children_right = tree.children_right")
    print(children_right)
    print("*************")

    # Create a list of nodes to remove
    nodes_to_remove = []
    for i in range(n_nodes):
        if children_left[i] == children_right[i]:
            n_leaves = 1
            for j in (children_left[i], children_right[i]):
                if j != -1:
                    n_leaves += 1
            if n_leaves <= min_leaves:
                nodes_to_remove.append(i)
    #Print the nodes to removed
    print('Nodes to be removed')
    print(nodes_to_remove)
    print("*************")
    # Remove the nodes
    for i in nodes_to_remove:
        children_left[i] = -1
        children_right[i] = -1

prune_tree_leaf_bound(clf.tree_, 2)


# evaluate the pruned tree on the test set
accuracy = clf.score(X_test, y_test)
print("Accuracy of the pruned tree: {:.2f}%".format(accuracy*100))
print("*************")

#Print the pruned tree
print("X_test, y_test:", clf.score(X_test, y_test))
print("*************")
print("X_train, y_train:", clf.score(X_train, y_train))
print("*************")
print("clf.tree_:", clf.tree_)
print("*************")
print("clf.tree_.node_count:", clf.tree_.node_count)
print("*************")
print("clf.tree_.children_left:", clf.tree_.children_left)
print("*************")
print("clf.tree_.children_right:", clf.tree_.children_right)
print("*************")
print("clf.tree_.feature:", clf.tree_.feature)
print("*************")
print("clf.tree_.threshold:", clf.tree_.threshold)
print("*************")
print("clf.tree_.value:", clf.tree_.value)
print("*************")
print("clf.tree_.impurity:", clf.tree_.impurity)
print("*************")
print("clf.tree_.n_node_samples:", clf.tree_.n_node_samples)
print("*************")
print("clf.tree_.weighted_n_node_samples:", clf.tree_.weighted_n_node_samples)
print("*************")
print("clf.tree_.max_depth:", clf.tree_.max_depth)
print("*************")
print("clf.tree_.max_n_classes:", clf.tree_.max_n_classes)
print("*************")
print("clf.tree_.n_classes:", clf.tree_.n_classes)
print("*************")
print("clf.tree_.n_outputs:", clf.tree_.n_outputs)
print("*************")
print("clf.tree_.n_features:", clf.tree_.n_features)
print("*************")
print("clf.tree_.max_features:", clf.tree_.n_features)
print("*************")
```
