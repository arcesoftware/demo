```python
import torch

torch.pi = torch.acos(torch.zeros(1)) * 2

epsilon = 0.1
torch.manual_seed(8675309)

xs = torch.randn(size=(300,))
ys = []
for i in range(len(xs)):
    numerator = torch.sum(torch.exp(-(xs[:i] - xs[i])**2 / (2 * epsilon**2)))
    denominator = torch.sqrt(2*torch.pi*epsilon**2) * len(xs)
    ys.append(numerator / denominator)

xd = torch.arange(torch.min(xs), torch.max(xs), 0.01)
yd = torch.exp(-xd**2/2) / torch.sqrt(2 * torch.pi)

plt.plot(xd, yd, 'x', label='density')
plt.scatter(xs, ys)
plt.axvline(x=0)
plt.axvline(x=torch.mean(xs), linestyle='--', color='purple')
plt.title(f'sample mean: {float(torch.mean(xs).item()):.2f}')
plt.legend()
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt

pi = torch.acos(torch.zeros(1)) * 2

epsilon = 0.1
torch.manual_seed(8675309)

xs = torch.randn(size=(300,))
ys = []
for i in range(len(xs)):
    numerator = torch.sum(torch.exp(-(xs[:i] - xs[i])**2 / (2 * epsilon**2)))
    denominator = torch.sqrt(2*pi*epsilon**2) * len(xs)
    ys.append(numerator / denominator)

xd = torch.arange(torch.min(xs), torch.max(xs), 0.01)
yd = torch.exp(-xd**2/2) / torch.sqrt(2 * pi)

plt.plot(xd, yd, 'x', label='density')
plt.scatter(xs, ys)
plt.axvline(x=0)
plt.axvline(x=torch.mean(xs), linestyle='--', color='purple')
plt.title(f'sample mean: {float(torch.mean(xs).item()):.2f}')
plt.legend()
plt.show()
```
