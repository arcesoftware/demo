```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def rand_unit() -> float:
    return random.random()

def rand_interval(a: float, b: float) -> float:
    return a + (b - a) * rand_unit()

def polar_coordinates(wx: float, wy: float) -> tuple:
    r = (wx - cx) ** 2 + (wy - cy) ** 2
    theta = math.atan2(wy - cy, wx - cx)
    return r, theta

def inverse_mapping(w: float) -> float:
    return math.sqrt(w)

def polar_coordinates(wx: float, wy: float, cx: float, cy: float) -> tuple:
    r = (wx - cx) ** 2 + (wy - cy) ** 2
    theta = math.atan2(wy - cy, wx - cx)
    return r, theta

def inverse_step(wx: float, wy: float, grid: np.ndarray, cx: float, cy: float, Nmax: int, res: int, depth: int, maxdepth: int):
    if depth >= maxdepth:
        return

    r, theta = polar_coordinates(wx, wy, cx, cy)
    wx = inverse_mapping(r) * math.cos(theta / 2)
    wy = inverse_mapping(r) * math.sin(theta / 2)

    xpos, ypos, xneg, yneg = grid_coordinates(wx, wy, res)
    
    if 0 <= xpos < res and 0 <= ypos < res and grid[xpos][ypos] < Nmax:
        grid[xpos][ypos] += 1
        inverse_step(wx, wy, grid, cx, cy, Nmax, res, depth + 1, maxdepth)
    
    if 0 <= xneg < res and 0 <= yneg < res and grid[xneg][yneg] < Nmax:
        grid[xneg][yneg] += 1
        inverse_step(-wx, -wy, grid, cx, cy, Nmax, res, depth + 1, maxdepth)

def main():
    while True:
        try:
            cx_str = input("Enter c (real and imaginary parts ie: -0.8 + 0.156): ")
            # Extracting real and imaginary parts
            cx_parts = cx_str.split()
            cx = float(cx_parts[0])
            cy = float(cx_parts[2])
            break
        except (ValueError, IndexError):
            print("Invalid input format. Please enter real and imaginary parts separated by space.")

    Nmax = int(input("Enter maximum number of points per grid box: ie:10 "))
    res = int(input("Enter grid resolution: ie: 800 "))
    maxdepth = int(input("Enter maximum depth of tree search: ie:10 "))

    assert Nmax > 0
    assert res >= 10
    assert res % 2 == 0
    assert maxdepth > 0

    grid = np.zeros((res, res))

    wx = np.random.normal(0, 0.1618033)
    wy = np.random.normal(0, 0.1618033)

    inverse_step(wx, wy, grid, cx, cy, Nmax, res, 0, maxdepth)

    plt.imshow(grid, cmap='hot')
    plt.show()

    plt.imsave('fractal_image.png', grid, cmap='hot')

if __name__ == '__main__':
    main()
```

```python
from juliart.main import JuliaSet
import matplotlib.pyplot as plt
import cv2

def generate_julia_set(resolution=1000, color="random", theme="random", rgb=None, iterations=200, zoom=1.0):
    julia_set = JuliaSet(resolution=resolution, color=color, theme=theme, rgb=rgb, iterations=iterations)
    julia_set.generate_image(zoom=zoom)
    return julia_set

def save_image(julia_set, filename):
    julia_set.save_image(filename)

def plot_image(filename):
    img = cv2.imread(filename)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

def main():
    # Generate Julia set
    julia_set = generate_julia_set(resolution=1000, color="random", theme="random", iterations=200, zoom=1.5)
    
    # Save Julia set image
    save_image(julia_set, "myimage.png")
    
    # Plot Julia set image
    plot_image("myimage.png")

if __name__ == "__main__":
    main()
```
